```python
import os, sys

filepath = os.path.abspath('../')

if filepath not in sys.path:
    sys.path.append(filepath)
```

```python
# import data

from src import Train

valueMapping, (X_likeability_test, y_likeability_test), (X_price_test, y_price_test) = Train.TrainAndSaveModels()
```

Classification and Regression Models Test

```python
import sys_dir as sd
import joblib

# classification models

classificationModels = {'DTCM' : joblib.load(sd.DTCM), 'KNNC' : joblib.load(sd.KNNC), 'LogRM' : joblib.load(sd.LogRM)}

classificationPredictions = {}

for modelName, model in classificationModels.items():
    classificationPredictions[modelName] = model.predict(X_likeability_test)

classificationResults= {} # use this for eval

for modelName, predictions in classificationPredictions.items():
    data = X_likeability_test.copy()
    data['true value'] = y_likeability_test
    data['predictions'] = predictions
    classificationResults[modelName] = data


# regression models
regressionModels = {'DTRM' : joblib.load(sd.DTRM), 'KNNR' : joblib.load(sd.KNNR), 'LinRM' : joblib.load(sd.LinRM)}

regressionPredictions = {}
regressionResults = {} # use this for eval

for modelName, model in regressionModels.items():
    regressionPredictions[modelName] = model.predict(X_price_test)

for modelName, predictions in regressionPredictions.items():
    data = X_price_test.copy()
    data['true value'] = y_price_test
    data['predictions'] = predictions
    regressionResults[modelName] = data
```

Classification Models Evaluation

Decision Tree Classifier Model

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import pandas as pd

print('Data Frame: ')
display(classificationResults['DTCM'])

trueValueDTCM = classificationResults['DTCM']['true value']
predictionsDTCM = classificationResults['DTCM']['predictions']

print('Confusion Matrix: ')
display(pd.DataFrame(confusion_matrix(trueValueDTCM, predictionsDTCM),index=['True 1', 'True 0'], columns=['Pred 1', 'Pred 0']))

print(f'Accuracy: {accuracy_score(trueValueDTCM, predictionsDTCM)}')
print(f'Precision Score: {precision_score(trueValueDTCM, predictionsDTCM)}')
print(f'Recall Score: {recall_score(trueValueDTCM, predictionsDTCM)}')
print(f'F1 Score: {f1_score(trueValueDTCM, predictionsDTCM)}')

```

```python
print('Data Frame: ')
display(classificationResults['KNNC'])

trueValueKNNC = classificationResults['KNNC']['true value']
predictionsKNNC = classificationResults['KNNC']['predictions']

print('Confusion Matrix: ')
display(pd.DataFrame(confusion_matrix(trueValueKNNC, predictionsKNNC),index=['True 1', 'True 0'], columns=['Pred 1', 'Pred 0']))

print(f'Accuracy: {accuracy_score(trueValueKNNC, predictionsKNNC)}')
print(f'Precision Score: {precision_score(trueValueKNNC, predictionsKNNC)}')
print(f'Recall Score: {recall_score(trueValueKNNC, predictionsKNNC)}')
print(f'F1 Score: {f1_score(trueValueKNNC, predictionsKNNC)}')
```
